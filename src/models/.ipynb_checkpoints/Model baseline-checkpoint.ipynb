```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import confusion_matrix as cmat

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, log_loss
from sklearn.preprocessing import StandardScaler
```

```python
np.random.seed(13)
```

```python
game_stats = pd.read_csv("data/csv/game_teams_stats.csv")
```

```python
game_stats.head()
```

```python
home = game_stats.loc[game_stats['HoA']=='home']
away = game_stats.loc[game_stats['HoA']=='away']
```

```python
game_stats = home.merge(away, on='game_id', suffixes=['_home', '_away'])
```

```python
game_stats.head()
```

```python
print(game_stats.columns)
```

```python
relevant_columns = ['game_id', 'won_home', 'settled_in_home', 'shots_home', 'hits_home', 'pim_home',
                   'powerPlayOpportunities_home', 'faceOffWinPercentage_home', 'giveaways_home',                            'takeaways_home', 'shots_away', 'hits_away', 'pim_away',                                                'powerPlayOpportunities_away', 'giveaways_away', 'takeaways_away']
```

```python
game_stats = game_stats.loc[:, relevant_columns]
```

```python
game_stats.head()
```

```python
def result(x):
    if x[1] == "OT":
        return "b_tie"
    elif x[0]:
        return "a_home_win"
    else:
        return "c_away_win"
```

```python
game_stats['result'] = game_stats.loc[:,['won_home', 'settled_in_home']].apply(result, axis=1)
```

```python
game_stats.head()
```

```python
id_train = np.load("data/numpy/id_train.npy")
id_test = np.load("data/numpy/id_test.npy")

X_train = game_stats.loc[game_stats['game_id'].isin(id_train), 
                        'shots_home':'takeaways_away'].values
y_train = game_stats.loc[game_stats['game_id'].isin(id_train), 
                        'result'].astype('category').cat.codes

X_test = game_stats.loc[game_stats['game_id'].isin(id_test), 
                        'shots_home':'takeaways_away'].values
y_test = game_stats.loc[game_stats['game_id'].isin(id_test), 
                        'result'].astype('category').cat.codes
```

```python
scaler = StandardScaler().fit(X_train)

X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
model_logistic = LogisticRegression(random_state=13).fit(X_train, y_train)
```

```python
print(model_logistic.coef_.shape)
print(model_logistic.intercept_)
```

```python
print(model_logistic.score(X_test, y_test))
```

```python
y_pred_logistic = model_logistic.predict(X_test)
```

```python
print(np.sum(np.equal(y_pred_logistic, 0)))
print(np.sum(np.equal(y_pred_logistic, 1)))
print(np.sum(np.equal(y_pred_logistic, 2)))
```
